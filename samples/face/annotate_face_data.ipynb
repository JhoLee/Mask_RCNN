```python
import os
import random
import sys

import tqdm
import utils as face_utils
```

```python
ROOT_DIR = os.path.realpath('../..')

DATASET_DIR = os.path.join('.', 'face_data')
TRAIN_DIR = os.path.join(DATASET_DIR, 'train')
VAL_DIR = os.path.join(DATASET_DIR, 'val')

TRAIN_MASK_DIR = os.path.join(TRAIN_DIR, 'mask')
TRAIN_ORIGINAL_DIR = os.path.join(TRAIN_DIR, 'original')
TRAIN_JSON_DIR = os.path.join(TRAIN_DIR, 'json')

VAL_MASK_DIR = os.path.join(VAL_DIR, 'mask')
VAL_ORIGINAL_DIR = os.path.join(VAL_DIR, 'original')
VAL_JSON_DIR = os.path.join(VAL_DIR, 'json')

t_o_list = os.listdir(TRAIN_ORIGINAL_DIR)
t_m_list = os.listdir(TRAIN_MASK_DIR)


v_o_list = os.listdir(VAL_ORIGINAL_DIR)
v_m_list = os.listdir(VAL_MASK_DIR)


print("Train Original Length: ", len(t_o_list))
print("Train Mask Length: ", len(t_m_list))
print("Val Original Length: ", len(v_o_list))
print("Val Mask Length: ", len(v_m_list))
```

```python
# Exract annotaion from 'train mask data'
for item in tqdm.tqdm(t_m_list):
    data = face_utils.load_info_from_mask(TRAIN_MASK_DIR, item, classes=['face'], values=[128])
    face_utils.extend_info_into_json(data, os.path.join(TRAIN_JSON_DIR, item.replace('bmp', 'json')))
```

```python
# Extract annotation from 'val mask data'
for item in tqdm.tqdm(v_m_list):
    data = face_utils.load_info_from_mask(VAL_MASK_DIR, item, classes=['face'], values=[128])
    face_utils.extend_info_into_json(data, os.path.join(VAL_JSON_DIR, item.replace('bmp', 'json')))
```

```python
t_j_list = os.listdir(TRAIN_JSON_DIR)
v_j_list = os.listdir(VAL_JSON_DIR)
print("Train Json Length: ", len(t_j_list))
print("Val Json Length: ", len(v_j_list))
```

```python
# Merge json files into one file - 'train mask data'
import json 


t_j_path = os.path.join(TRAIN_ORIGINAL_DIR, 'region_data.json')
face_utils.extend_info_into_json({}, t_j_path)

with open(t_j_path, 'r', encoding="utf-8") as f:
    _data = json.load(f)

for data in tqdm.tqdm(t_j_list):
    _t_j_path = os.path.join(TRAIN_JSON_DIR, data)
    with open(_t_j_path, 'r', encoding="utf-8") as f:
        _data.update(json.load(f))

face_utils.extend_info_into_json(_data, t_j_path)
```
